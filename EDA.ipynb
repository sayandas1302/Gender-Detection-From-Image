```python
# detection of eyes in the image

# loading the pre-trained  eye cascade classifier
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Load the image
img = cv2.imread('./dataset/initial faces/1_0_2_20161219205107772.jpg.chip.jpg')

# Convert the image to rgb
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Perform eye detection
eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=10)

# Iterate over detected eyes
eye_count = 0
for (ex, ey, ew, eh) in eyes:
    # Draw a rectangle around the eyes (within the face)
    cv2.rectangle(img_rgb, (ex, ey), (ex+ew, ey+eh), (255, 0, 0), 2)
    eye_count += 1

# Display the image with detections
plt.imshow(img_rgb)
plt.axis('off')
plt.title(f'Eyes Detected: {eye_count}')
```
