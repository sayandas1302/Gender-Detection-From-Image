### Importing Dependencies

```python
import pickle
import math
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, r2_score, mean_absolute_error
from tensorflow.keras import Sequential, layers, regularizers
```

### Calling the dataset

```python
with open('./artifacts/X.pkl', 'rb') as file:
    X = pickle.load(file)

with open('./artifacts/y_age.pkl', 'rb') as file:
    y_age = pickle.load(file).astype(float)    

with open('./artifacts/y_gender.pkl', 'rb') as file:
    y_gender = pickle.load(file).astype(int)
```

```python
# scaling the input array
X = X/255
```

### Model for Gender

```python
### train test split
X_train, X_test, y_train, y_test = train_test_split(X, y_gender, test_size=0.2)
```

```python
# model building 
gen_cnn_model = Sequential([
    layers.Conv2D(filters=20, kernel_size=(10, 10), activation='relu', input_shape=(100, 100, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(filters=5, kernel_size=(5, 5), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),

    layers.Dense(200, activation='sigmoid'),
    layers.Dense(200, activation='sigmoid'),
    layers.Dense(200, activation='sigmoid'),
    layers.Dense(200, activation='sigmoid'),
    layers.Dense(2, activation='softmax')
])

gen_cnn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

gen_cnn_model.fit(X_train, y_train, epochs=10)
```

```python
# prediction 
y_pred = [x.argmax() for x in gen_cnn_model.predict(X_test)]
y_pred
```

```python
# testing accuracy
print(classification_report(y_test, y_pred))
```

```python
# saving the model for gender
gen_cnn_model.save('./artifacts/gen_model.h5')
```
